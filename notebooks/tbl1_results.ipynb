# Table 1. Neuron-based classification results in 4-way evaluation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.chdir('..')
```

```python
neuron_df = pd.read_csv('experiments/trials/results/neuron_classification.csv')
neuron_4way_df = neuron_df[(neuron_df['num_img_per_trial'] == 4) & (neuron_df['model'] != 'cvcl-random')]
neuron_4way_df
```

```python
vanilla_df = pd.read_csv('experiments/trials/results/baseline.csv')
vanilla_4way_df = vanilla_df[vanilla_df['num_img_per_trial'] == 4]
vanilla_4way_df
```

```python
# Function to format accuracy with standard deviation
def format_acc(group):
    mean = group['overall_accuracy'].mean() * 100
    std = group['overall_accuracy'].std() * 100
    return f"{mean:.2f}{std:.2f}%"

def cal_class_acc(df):
    # Group by model and class_type, then aggregate
    result = df.groupby(['model', 'class_type']).apply(format_acc).unstack()

    # Rename columns for clarity
    result.columns.name = None
    result = result.rename(columns={'unseen': 'Out-of-vocab', 'seen': 'In-vocab', 'full': 'All'})

    # Reorder columns
    result = result[['In-vocab', 'Out-of-vocab', 'All']]

    # Reset index to make 'model' a column
    result = result.reset_index()

    # Rename the 'model' column
    result = result.rename(columns={'model': 'Model'})

    # Display the result
    return result
```

```python
vanilla_results = cal_class_acc(vanilla_4way_df)
print(vanilla_results)
```

```python
neuron_result = cal_class_acc(neuron_4way_df)
print(neuron_result)
```
